Voici un exercice de kata simple (difficulté estimée à 2/10) appelé "Le FizzBuzz Inversé" (ou Reverse FizzBuzz).

Objectif : Écrire une fonction reverse_fizzbuzz qui prend en paramètre une chaîne de caractères ("Fizz", "Buzz", "FizzBuzz" ou un nombre sous forme de chaîne comme "7") et retourne l'entier correspondant selon les règles classiques du FizzBuzz. Si la chaîne est "Fizz", retournez 3. Si c'est "Buzz", retournez 5. Si c'est "FizzBuzz", retournez 15. Pour tout autre nombre sous forme de chaîne (ex: "7"), retournez sa valeur entière (7).

Voici uniquement les codes de test pour valider votre implémentation en Go et en Python.

```python
import unittest
```

```python
def reverse_fizzbuzz(input: str) -> int:
    match input:
        case "Fizz":
            return 3
        case "Buzz":
            return 5
        case "FizzBuzz":
            return 15
        case _:
            return int(input)
```

```python
def test_reverse_fizzbuzz():
    assert reverse_fizzbuzz("Fizz") == 3
    assert reverse_fizzbuzz("Buzz") == 5
    assert reverse_fizzbuzz("FizzBuzz") == 15
    print('ok')
```

```python
class TestReverseFizzBuzz(unittest.TestCase):
    def test_fizz(self):
        self.assertEqual(reverse_fizzbuzz("Fizz"), 3)

    def test_buzz(self):
        self.assertEqual(reverse_fizzbuzz("Buzz"), 5)

    def test_fizzbuzz(self):
        self.assertEqual(reverse_fizzbuzz("FizzBuzz"), 15)

    def test_number(self):
        self.assertEqual(reverse_fizzbuzz("7"), 7)
        self.assertEqual(reverse_fizzbuzz("101"), 101)
```

```python
test_reverse_fizzbuzz()
```
